```
%tensorflow_version 1.x

import tensorflow as tf
import numpy as np
from datetime import datetime
import os

working_dir = 'DSC_project'

colab_default_dir = '/content/gdrive/My Drive/Colab Notebooks/'

original_dir = os.getcwd()

try:

  os.chdir(colab_default_dir)

  if not os.path.exists(working_dir):
    os.mkdir(working_dir)

  os.chdir(working_dir)
  print("current dir = ", os.getcwd())

except Exception as err:
  
  os.chdir(original_dir)
  print(str(err))
```

```
try:
    data = np.loadtxt('./삼대중량_인공지능_데이터(14).csv', delimiter=',')
    
    print("data.shape = ", data.shape, ", data.ndim = ", data.ndim)
    print(data)
    
except Exception as err:
    print(str(err))
```

```
xdata = data[ :  , 1:-1]
tdata = data[ :  , [-1]]

print("xdata.shape = ", xdata.shape, ", xdata.ndim = ", xdata.ndim)
print("tdata.shape = ", tdata.shape, ", tdata.ndim = ", tdata.ndim)

print("xdata = ", xdata)
print("tdata = ", tdata)
```

```
X = tf.placeholder(tf.float32, [None, 3])
T = tf.placeholder(tf.float32, [None, 1])

W = tf.Variable(tf.random_normal([3, 1]))
b = tf.Variable(tf.random_normal([1]))
```

```
y = tf.matmul(X, W) + b

loss = tf.reduce_mean(tf.square(T - y))
```

```
learning_rate = 1e-1

optimizer = tf.train.AdamOptimizer(learning_rate)

train = optimizer.minimize(loss)
```

```
with tf.Session() as sess:
  sess.run(tf.global_variables_initializer())
    
  start_time = datetime.now()
    
  for step in range(10001):
    loss_val, _ = sess.run([loss, train], feed_dict={X: xdata, T: tdata})
        
    if step % 1000 == 0:
      print("step = ", step, ", loss_val = ", loss_val)
            
  end_time = datetime.now()

  print("")
  print("Elapsed Time => ", end_time - start_time)

  print("\nW = ", sess.run(W))
  print("b = ", sess.run(b))

  predict_val = sess.run(y, feed_dict={X: [[70., 100., 80.]]})

  print("\npredict_val = ", predict_val)
```
